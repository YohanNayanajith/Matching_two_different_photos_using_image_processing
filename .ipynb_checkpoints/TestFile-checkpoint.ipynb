```python
import cv2
import numpy as np

img = cv2.imread("simpsons.jpg") #coloured format
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #gray scale image

#template = cv2.imread("barts_face.jpg")
#gray_template = cv2.cvtColor(template,cv2.COLOR_BGR2GRAY)
template = cv2.imread("barts_face.jpg",cv2.IMREAD_GRAYSCALE)
w,h = template.shape[::-1] #rows = height,columns = width

result = cv2.matchTemplate(gray_img,template,cv2.TM_CCOEFF_NORMED)
#print(result)
loc = np.where(result >= 0.9)
#print(loc)
for pt in zip(*loc[::-1]):
    #print(pt)
    cv2.rectangle(img,pt,(pt[0]+w,pt[1]+h),(0,255,0),3)

cv2.imshow("Image",img)
cv2.imshow("Result",result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
import cv2

video_capture = cv2.VideoCapture(0)
template = cv2.imread("image.jpg",cv2.IMREAD_GRAYSCALE)
w,h = template.shape[::-1]

while True:
    _,frame = video_capture.read()
    gray_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    res = cv2.matchTemplate(gray_frame,template,cv2.TM_CCOEFF_NORMED)
    location = np.where(res>=0.3)
    
    for p in zip(*location[::-1]):
        cv2.rectangle(frame,p,(p[0]+w,p[1]+h),(0,0,255),3)
    
    cv2.imshow("Video",frame)
    key = cv2.waitKey(1)
    if(key == 27):
        break
        
video_capture.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread("simpsons.jpg") #coloured format
gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) #gray scale image
h1,w1 = gray_img.shape
final_img = img

template = cv2.imread("barts_face.jpg",cv2.IMREAD_GRAYSCALE)
w,h = template.shape[::-1] #rows = height,columns = width

result = cv2.matchTemplate(gray_img,template,cv2.TM_CCOEFF_NORMED)
loc = np.where(result >= 0.9)

for pt in zip(*loc[::-1]):
    global coordinates
    coordinates = pt
    cv2.rectangle(img,pt,(pt[0]+w,pt[1]+h),(0,255,0),3)

crop_width_size = 3
crop_height_size = 3
height = int(h1 / crop_height_size)
width = int(w1 / crop_width_size )

correct_rect = False
rect_number = 0.9
for ih in range(crop_height_size):
    for iw in range(crop_width_size):
        x = int(w1/crop_width_size * iw) 
        y = int(h1/crop_height_size * ih)

        cv2.rectangle(final_img,(x,y),(x+width,y+height),(0,0,255),1)
        if((coordinates[0] >= x and coordinates[0] <= (x+width)) and (coordinates[1] >= y and coordinates[1] <= (y+height))):
            if(((coordinates[0]+int(w/2)) >= x and (coordinates[0]+int(w/2)) <= (x+width)) and ((coordinates[1]+h) >= y and (coordinates[1]+h) <= (y+height))):
                correct_rect = True
                rect_number = iw+1
        else:
            if(((coordinates[0]+int(w/2)) >= x and (coordinates[0]+int(w/2)) <= (x+width)) and ((coordinates[1]+h) >= y and (coordinates[1]+h) <= (y+height))):
                correct_rect = True
                rect_number = iw+1

print(coordinates)
if(correct_rect == True):
    print(rect_number)
cv2.imshow("Image",img)
#cv2.imshow("Final Image",final_img)
#cv2.imshow("Result",result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Feature Detection (SIFT,SURF,ORB)

### SURF algorithm is faster than SIFT
    SURF algorithm ekata pay karanna one commercial use ekakadi

### ORB algo eke default detect karaganne features 500k vitharai

```python
import cv2
import numpy as np

img = cv2.imread("simpsons.jpg",cv2.IMREAD_GRAYSCALE)

#sift = cv2.xfeatures2d.SIFT_create()
#surf = cv2.xfeatures2d.SURF_create()
orb = cv2.ORB_create(nfeatures=1500)

#kp = sift.detect(img,None) #image eke keypoints detect karagannava
#key_points,descriptors = sift.detectAndCompute(img,None) 
#key_points,descriptors = surf.detectAndCompute(img,None) 
key_points,descriptors = orb.detectAndCompute(img,None) 

#img = cv2.drawKeypoints(img,kp,None)
img = cv2.drawKeypoints(img,key_points,None)
#print(descriptors)

cv2.imshow("Image",img)
key = cv2.waitKey(0)
cv2.destroyAllWindows()
```
